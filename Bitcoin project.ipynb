```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
import plotly.express as px
import yfinance as yf
from autots import AutoTS
```

```python
data = yf.download("BTC-USD",
                   start ='2014-01-01',
                   end = '2022-04-04',
                   progress = False)
df = pd.DataFrame(data)
df.head(10)
```

```python
dfres = df.reset_index()
dfres
```

```python
dfres = dfres[["Date","Close"]]
dfres["Date"]= pd.to_datetime(dfres.Date)
```

```python
px.line(dfres["Close"],x=dfres["Date"],y=dfres["Close"],title = "Bitcoin price")
```

```python
model = AutoTS(forecast_length=30, frequency='infer', 
               ensemble='simple', drop_data_older_than_periods=200)
model = model.fit(dfres, date_col='Date', value_col='Close', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print("Price prediction of Bitcoin")
print(forecast)
```
